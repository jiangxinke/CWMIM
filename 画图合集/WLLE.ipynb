```python
import numpy as np
from lib.get_manifold import get_color
from lib.WLLE import WLLE
import matplotlib.pyplot as plt
```

```python
cho = ['YC01_rel', 'YC02_rel', "DDH_left", "DDH_right"]
data = cho[2]
file = r"D:\projects\Datasets\Original\{}.csv".format(data)
```

```python
X = np.loadtxt(fname=file, skiprows=1, delimiter=",")[:, 0:3]
```

```python
c = get_color(data)
```

```python
for i in [0, 1, 2]:
    h = X[:, i]
    temp = (h - h.min()) / (h.max() - h.min())
    X[:, i] = temp
```

```python
mani = WLLE(X, n_neighbors=150, n_components=2, new=True, gamma=100)
```

```python
x = mani[:, 0]
y = mani[:, 1]

plt. clf()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5), tight_layout=True)
ax.set_xticks([])
ax.set_yticks([])
ax.scatter(x, y, s=5, c=c, marker='o', alpha=1)

# plt.savefig(r"D:\projects\Datasets\Embeddings\WLLE\left\{}.png".format(k))
plt.show()
```

```python
for i in np.arange(50, 500, 20):
    mani = WLLE(X, n_neighbors=i, n_components=2, new=True, gamma=100)
    x = mani[:, 0]
    y = mani[:, 1]

    plt. clf()
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5), tight_layout=True)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.scatter(x, y, s=5, c=c, marker='o', alpha=1)

    plt.savefig(r"D:\projects\Datasets\Embeddings\WLLE\left\{}.png".format(i))
#     plt.show()
```
